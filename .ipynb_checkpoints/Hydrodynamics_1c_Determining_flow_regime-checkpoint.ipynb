### Recall the catalyst parameters and the operating range of liquid and gas velocities obtained from part 1a and 1b where we have obtained the minimum fluidization velocity and the particle terminal velocities. 

Catalyst particle diameter: $d_p = 3.5 \times 10^{-3} \, [m]$

Catalyst density: $\rho_s =  1.72 \times 10^5 \, [\frac{kg}{m^3}]$

Liquid viscosity: $\mu_l = 1.7 \times 10^{-4} \, [Pa \cdot s] $ = $[\frac{kg}{m \cdot s}]$

Liquid density: $\rho_{l} = 730 \, [\frac{kg}{m^3}]$

Gravitational acceleration: $g = 9.8 \, [\frac{m}{s^2}]$

### Recall the optimal column diameter which we have simulated in the overall material balance


Column Diameter: $D_c = 0.2 [m]$


```python
# Catalyst particle diameter [m]
dp = 3.5e-3  

# Catalyst Density [kg / m^3]
rho_s = 1.72e5  

# Liquid viscosity [Pa.s] = [kg/m/s]
mu_l = 1.7e-4

# Liquid Density [kg / m^3]
rho_l = 730  

# Gravitational acceleration in m/s^2
g = 9.8  

# Column Diameter [m]
Dc = 0.2 
```

## Thus we will use the correlations of flow regime transitions provided from [1] to plot a map to visualise the flow regime boundaries of our three-phase fluidized bed reactor.


### Onset of dispersed bubble flow regime

$$
\dfrac{U_g}{U_l} = 0.721 \, Fr_g^{0.339} \, Ar_l^{0.0746} \left(\dfrac{\rho_s}{\rho_l}\right)^{-0.667}
\tag{1}
$$


```python
def onset_of_dispersed_bubble(Ug, rho_s=1.72e5, rho_l=730):
    """
    Ul for the onset of dispersed bubble flow regime based on inputted Ug
    """
    Frg = calculate_Frg(Ug)
    
    Arl = calculate_Arl()
    
    numer = Ug
    denom = 0.721 * (Frg**(0.339)) * (Arl**(0.0746)) * (rho_s/rho_l)**(-0.667)
    
    return numer / denom
```

### Superficial liquid velocity, $U_l$, is low when the following holds: 

$$
U_l \leq \left[8.82 \, \left(\dfrac{\rho_l \, d_p}{\mu_l} \right)^{-0.376} \, (g \, D_c)^{0.312} \, Ar_l^{0.419} \, \left(\dfrac{D_c}{d_p}\right)^{0.744} \right]
\tag{2}
$$

### Onset of slug flow regime (For low superficial liquid velocity)

$$
\dfrac{U_g}{U_l} = 6.97 \times 10^5 \, Re_l^{-0.918} \, Ar_l^{0.805} \left(\dfrac{D_c}{d_p}\right)^{1.01}
\tag{3}
$$

### Onset of slug flow regime (For high superficial liquid velocity)

$$
\dfrac{U_g}{U_l} = 20.3 \, Fr_{gD}^{0.302} \, Ar_l^{-0.0861} \left(\dfrac{D_c}{d_p}\right)^{-0.318}
\tag{4}
$$


```python
def onset_of_slug(Ug, dp=3.5e-3, rho_l=730, rho_s=1.72e5, Dc=0.2, mu_l=1.7e-4, g=9.8):
    """
    Ul for the onset of slug flow regime based on inputted Ug
    """
    
    def low_Ul(Ug):
        """
        Low superficial liquid velocity Ul for the onset of slug flow regime
        """
        
        Arl = calculate_Arl()
        numer = Ug
        denom = 6.97e5 * ((dp * rho_l / mu_l)**(-0.918)) * (Arl**(0.805)) * ((Dc / dp)**(1.01))
        result = (numer / denom)**(500/41)
        return result
    
    
    def high_Ul(Ug):
        """
        High superficial liquid velocity Ul for the onset of slug flow regime
        """
        
        FrgD = calculate_FrgD(Ug)
        Arl = calculate_Arl()
        numer = Ug
        denom = 20.3 * (FrgD**(0.302)) * (Arl**(-0.0861)) * ((Dc / dp)**(-0.318))
        return numer / denom
    
    
    def criteria(Ul):
        
        Arl = calculate_Arl()
        limit = 8.82 * (((rho_l * dp) / mu_l)**(-0.376))*((g * Dc)**0.312) * (Arl**(0.419))*((Dc / dp)**(0.744))
        
        if (Ul <= limit):
            return low_Ul(Ug)
        return high_Ul(Ug)
    
    return criteria(high_Ul(Ug))
```

### Onset of churn flow regime

$$
\dfrac{U_g}{U_l} = 53.4 \, Re_l^{-1.3} \, Ar_l^{0.468} \left(\dfrac{\rho_s}{\rho_l}\right)^{0.454}
\tag{5}
$$

```python
def onset_of_churn(Ug, rho_l=730, rho_s=1.72e5):
    """
    Ul for the onset of churn flow regime
    """
    
    Arl = calculate_Arl()
    numer = Ug
    denom = 53.4 * ((dp * rho_l / mu_l)**(-1.3)) * (Arl**(0.468)) * ((rho_s/ rho_l)**(0.454))
    result = (numer / denom)**(0.3)
    return result
```

### Onset of bridging flow regime

$$
\dfrac{U_g}{U_l} = 219 \, Re_{l}^{-1.24} \, Ar_l^{0.367} \left(\dfrac{\rho_s}{\rho_l}\right)^{0.444}
\tag{6}
$$

```python
def onset_of_bridging(Ug, rho_l=730, rho_s=1.72e5):
    """
    Ul for the onset of bridging flow regime.
    """
    Arl = calculate_Arl()
    numer = Ug
    denom = 219 * ((dp * rho_l / mu_l)**(-1.24)) * (Arl**(0.367)) * ((rho_s/ rho_l)**(0.444))
    result = (numer / denom)**(0.24)
    return result    
```

where

$$
Ar_l = \dfrac{d_p^3 \, \rho_l \, (\rho_s - \rho_l) \, g}{\mu_l^2}
\hspace{1cm} \text{(Liquid Archimedes Number)}
$$

$$
Fr_g = \dfrac{U_g^2}{g \, d_p}
\hspace{1cm} \text{(Gas Froude Number)}
$$

$$
Fr_{gD} = \dfrac{U_g^2}{g \, D_c}
\hspace{1cm} \text{(Gas Froude Number based on column diameter)}
$$

$$
Re_l = \dfrac{U_l \, d_p \, \rho_l}{\mu_l}
\hspace{1cm} \text{(Particle Reynolds Number based on superficial liquid velocity)}
$$

```python
def calculate_Arl(dp=3.5e-3, rho_l=730, rho_s=1.72e5, mu_l=1.7e-4, g=9.8):
    """Liquid Archimedes Number"""
    
    numer = dp**3 * rho_l * (rho_s - rho_l) * g
    denom = mu_l**2
    
    return numer / denom

def calculate_Frg(Ug, dp=3.5e-3, g=9.8):
    """Gas Froude Number"""
    
    numer = Ug**2
    denom = g * dp
    
    return numer / denom

def calculate_FrgD(Ug, Dc=0.2, g=9.8):
    """Gas Froude Number based on column diameter"""
    
    numer = Ug**2
    denom = g * Dc
    
    return numer / denom

# def calculate_Rel(Ul, dp=3.5e-3, rho_l=730, mu_l=1.7e-4):
#     """Particle Reynolds Number based on Superficial Liquid velocity"""
    
#     numer = Ul * dp * rho_l
#     denom = mu_l
    
#     return numer / denom
```

```python
# Superficial gas velocity range [m/s]
Ug_range = np.linspace(0.1, 5, 100)  

Ul_dispersed_boundary = np.zeros_like(Ug_range)

Ul_slug_boundary = np.zeros_like(Ug_range)

Ul_churn_boundary = np.zeros_like(Ug_range)

Ul_bridging_boundary = np.zeros_like(Ug_range)
```

```python
# Calculate boundaries based on equations
for i, Ug in enumerate(Ug_range):
    
    # Onset of dispersed bubble flow regime
    Ul_dispersed_boundary[i] = onset_of_dispersed_bubble(Ug)

    # Onset of slug flow regime (both low Ul and high Ul)
    Ul_slug_boundary[i] = onset_of_slug(Ug)

    # Onset of churn flow regime
    Ul_churn_boundary[i] = onset_of_churn(Ug)

    # Onset of bridging flow regime
    Ul_bridging_boundary[i] = onset_of_bridging(Ug)

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(Ug_range, Ul_dispersed_boundary, label='Dispersed bubble flow regime')
plt.plot(Ug_range, Ul_slug_boundary, label='Slug flow regime')
plt.plot(Ug_range, Ul_churn_boundary, label='Churn flow regime')
plt.plot(Ug_range, Ul_bridging_boundary, label='Bridging flow regime')

plt.xlabel('Superficial Gas Velocity, $U_g$ (m/s)')
plt.ylabel('Superficial Liquid Velocity, $U_l$ (m/s)')
plt.title('Flow Regime Map of a Three-Phase Fluidized Bed Reactor')
plt.legend()
plt.grid(True)

plt.savefig("./graphs/Flow_Regime_Map.png")
plt.show()
```

# We can see that since our minimum fluidization liquid velocity is $ \sim 5 [\frac{m}{s}]$ such that we are definitely operating at dispersed bubble flow regime where the ideal operating superficial gas velocity is at $\sim 3 [\frac{m}{s}]$.
